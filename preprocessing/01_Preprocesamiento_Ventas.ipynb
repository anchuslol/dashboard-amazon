# Limpiar columnas

```python
# Definir funci√≥n para limpiar porcentajes si no existe
def limpiar_porcentaje(serie):
    """
    Convierte strings de porcentaje como '92.38%' o '7,30%' a float (0.9238 o 0.073).
    Maneja comas y espacios.
    """
    return (
        serie.astype(str)
        .str.replace('%', '', regex=False)
        .str.replace(',', '.', regex=False)
        .str.strip()
        .replace('', np.nan)
        .astype(float) / 100
    )

# Identificar columnas de porcentajes en df_asin (no df_asin_child)
columnas_porcentaje_asin = [col for col in df_asin.columns if "porcentaje" in col.lower()]

# Mostrar qu√© columnas se van a limpiar
print("Columnas de porcentajes encontradas en df_asin:")
for col in columnas_porcentaje_asin:
    print(f"  - {col}")

# Aplicar limpieza a las columnas de porcentajes en df_asin
for col in columnas_porcentaje_asin:
    print(f"Limpiando columna: {col}")
    print(f"Antes: {df_asin[col].head(3).tolist()}")
    df_asin[col] = limpiar_porcentaje(df_asin[col])
    print(f"Despu√©s: {df_asin[col].head(3).tolist()}")
    print("-" * 50)

# Verificar el resultado
print("\nVerificando tipos de datos:")
for col in columnas_porcentaje_asin:
    print(f"{col}: {df_asin[col].dtype}")
```

```python
# Definir funci√≥n para limpiar columnas monetarias
def limpiar_monetario(serie):
    return (
        serie.astype(str)
        .str.replace('.', '', regex=False)  # Quitar separador de miles
        .str.replace('‚Ç¨', '', regex=False)
        .str.replace('\xa0', '', regex=False)  # Quitar espacio no separable
        .str.replace(',', '.', regex=False)    # Cambiar coma decimal por punto
        .str.strip()
        .replace('', np.nan)
        .astype(float)
    )

# Identificar columnas monetarias en df_asin
columnas_monetarias_asin = [col for col in df_asin.columns if "venta" in col or "ventas" in col or "importe" in col or "precio" in col]

# Limpiar columnas monetarias
for col in columnas_monetarias_asin:
    df_asin[col] = limpiar_monetario(df_asin[col])
```

```python
# Definir funci√≥n para limpiar separadores de miles (comas)
def limpiar_separadores_miles(col):
    """
    Elimina separadores de miles (comas) y convierte a int
    Ejemplo: "1,234" ‚Üí 1234, "567" ‚Üí 567
    """
    return (
        col.astype(str)  # Asegurar que sea string
           .str.replace(",", "", regex=False)  # Quitar separadores de miles (comas)
           .str.strip()  # Quitar espacios
           .replace("", "0")  # Reemplazar vac√≠os por 0
           .astype(int)  # Convertir a entero
    )

# Columnas num√©ricas que pueden tener separadores de miles
columnas_numericas = [
    'unidades_encargadas',
    'unidades_encargadas_b2b',
    'sesiones',
    'sesiones_b2b',
    'vistas_de_pagina',
    'vistas_de_pagina_b2b',
    'sesiones_aplicaciones_de_dispositivos_moviles',
    'sesiones_aplicaciones_de_dispositivos_moviles__b2b',
    'sesiones_navegador',
    'sesiones_total',
    'visitas_de_pagina_aplicacion_de_dispositivo_movil',
    'visitas_total',
    'visitas_navegador'
]

# Aplicar limpieza a las columnas num√©ricas
for col in columnas_numericas:
    if col in df_asin.columns:
        print(f"Limpiando separadores de miles en: {col}")
        print(f"Antes: {df_asin[col].head(3).tolist()}")
        df_asin[col] = limpiar_separadores_miles(df_asin[col])
        print(f"Despu√©s: {df_asin[col].head(3).tolist()}")
        print("-" * 50)
    else:
        print(f"‚ö†Ô∏è Columna '{col}' no encontrada en el DataFrame")

# Verificar el resultado
print("\nüìä Resumen de las columnas num√©ricas limpiadas:")
for col in columnas_numericas:
    if col in df_asin.columns:
        print(f"{col}: {df_asin[col].dtype}")
```

```python
import re
import unicodedata

def limpiar_columna(col):
    # Eliminar acentos
    col = unicodedata.normalize('NFKD', col).encode('ascii', errors='ignore').decode('utf-8')
    # Reemplazar caracteres especiales por espacio
    col = re.sub(r'[\-:()\t]', ' ', col)
    # Reemplazar m√∫ltiples espacios por uno solo
    col = re.sub(r'\s+', ' ', col)
    # Quitar espacios al inicio y final
    col = col.strip()
    # Reemplazar espacios por guion bajo y pasar a min√∫sculas
    col = col.replace(' ', '_').lower()
    return col

df_pedidos.columns = [limpiar_columna(col) for col in df_pedidos.columns]
df_pedidos.head(2)
```

```python
# Funci√≥n completa para convertir fechas a hora local
import pandas as pd
import pytz
from datetime import datetime

def convertir_fechas_hora_local(df, col_marketplace='sales_channel'):
    """
    Convierte columnas de fecha a hora local seg√∫n el pa√≠s del marketplace.
    """
    
    # Mapeo de marketplaces a zonas horarias
    timezone_map = {
        'Amazon.es': 'Europe/Madrid',
        'Amazon.fr': 'Europe/Paris', 
        'Amazon.it': 'Europe/Rome',
        'Amazon.de': 'Europe/Berlin',
        'Amazon.nl': 'Europe/Amsterdam',
        'Amazon.se': 'Europe/Stockholm',
        'Amazon.co.uk': 'Europe/London',
        'Amazon.pl': 'Europe/Warsaw',
        'Amazon.com.tr': 'Europe/Istanbul',
        'Amazon.ie': 'Europe/Dublin',
        'Amazon.be': 'Europe/Brussels',
        'Amazon.com': 'America/New_York',
        'Amazon.ca': 'America/Toronto',
        'Amazon.com.mx': 'America/Mexico_City',
        'Amazon.co.jp': 'Asia/Tokyo',
        'Amazon.com.au': 'Australia/Sydney'
    }
    
    # Detectar columnas de fecha
    columnas_fecha = [col for col in df.columns if 'date' in col.lower() or 'fecha' in col.lower()]
    
    if not columnas_fecha:
        print("‚ö†Ô∏è No se detectaron columnas de fecha.")
        return df
    
    print(f"üïê Columnas de fecha detectadas: {columnas_fecha}")
    
    df_resultado = df.copy()
    
    for col_fecha in columnas_fecha:
        try:
            print(f"\nüîÑ Procesando columna: {col_fecha}")
            
            # Convertir a datetime
            df_resultado[col_fecha] = pd.to_datetime(df_resultado[col_fecha], errors='coerce')
            
            # Crear columnas adicionales √∫tiles
            df_resultado[f'{col_fecha}_solo_fecha'] = df_resultado[col_fecha].dt.date
            df_resultado[f'{col_fecha}_hora'] = df_resultado[col_fecha].dt.hour
            df_resultado[f'{col_fecha}_dia_semana'] = df_resultado[col_fecha].dt.day_name()
            df_resultado[f'{col_fecha}_mes'] = df_resultado[col_fecha].dt.month
            df_resultado[f'{col_fecha}_a√±o'] = df_resultado[col_fecha].dt.year
            
            # Crear tramos horarios
            df_resultado[f'{col_fecha}_tramo_horario'] = pd.cut(
                df_resultado[f'{col_fecha}_hora'],
                bins=[-1, 6, 12, 18, 24],
                labels=['Madrugada (0-6h)', 'Ma√±ana (6-12h)', 'Tarde (12-18h)', 'Noche (18-24h)']
            )
            
            print(f"‚úÖ {col_fecha} ‚Üí convertida a datetime + columnas derivadas")
            print(f"   Tipo final: {df_resultado[col_fecha].dtype}")
            
        except Exception as e:
            print(f"‚ùå Error procesando {col_fecha}: {e}")
    
    return df_resultado

# Aplicar la funci√≥n corregida a df_pedidos
if 'df_pedidos' in locals() and df_pedidos is not None and not df_pedidos.empty:
    print("üîÑ CONVIRTIENDO COLUMNAS DE FECHA EN df_pedidos")
    print("=" * 50)
    
    # Verificar columnas de fecha antes
    fecha_cols = [col for col in df_pedidos.columns if 'date' in col.lower()]
    print(f"Columnas con 'date': {fecha_cols}")
    
    if fecha_cols:
        for col in fecha_cols:
            print(f"Antes - {col}: {df_pedidos[col].dtype}")
        
        # Aplicar conversi√≥n
        df_pedidos = convertir_fechas_hora_local(df_pedidos, 'sales_channel')
        
        # Verificar despu√©s
        print("\nüìä RESULTADO:")
        for col in fecha_cols:
            print(f"Despu√©s - {col}: {df_pedidos[col].dtype}")
            print(f"   Muestra: {df_pedidos[col].head(2).tolist()}")
    else:
        print("‚ö†Ô∏è No se encontraron columnas con 'date'")
        print("Columnas disponibles:", df_pedidos.columns.tolist())
else:
    print("‚ö†Ô∏è df_pedidos no est√° disponible para procesar")
```

### INFORMES DE VENTAS

FUNCION GENERAL DE LIMPIEZA DE COLUMNAS, PORCENTAJES, MONETARIAS, NUMERICAS, FECHA

```python
import pandas as pd
import numpy as np
import unicodedata
import re

# 1. Normalizar nombres de columnas: quitar acentos y caracteres especiales, poner snake_case
def limpiar_columnas(df):
    new_cols = []
    for col in df.columns:
        col_no_accents = unicodedata.normalize('NFKD', str(col)).encode('ascii', errors='ignore').decode('utf-8')
        col_clean = re.sub(r'[-:()\t]', ' ', col_no_accents)
        col_clean = re.sub(r'\s+', ' ', col_clean).strip()
        col_snake = col_clean.replace(' ', '_').lower()
        new_cols.append(col_snake)
    df.columns = new_cols
    return df

# 2. Convertir porcentajes o ratios a proporciones decimales
def limpiar_porcentajes(df):
    porcentaje_cols = [col for col in df.columns if ('porcentaje' in col or 'ratio' in col)]
    for col in porcentaje_cols:
        df[col] = (df[col].astype(str)
                          .str.replace('%', '', regex=False)
                          .str.replace(',', '.', regex=False)
                          .str.strip())
        df[col] = pd.to_numeric(df[col], errors='coerce') / 100
    return df

# 3. Convertir columnas monetarias y promedios: eliminar el s√≠mbolo ‚Ç¨ y convertir a float
def limpiar_monetarios(df):
    monetarias = [c for c in df.columns if ('ventas' in c or 'promedio' in c 
                                            or 'importe' in c or 'precio' in c)]
    for col in monetarias:
        df[col] = (df[col].astype(str)
                        .str.replace('‚Ç¨','', regex=False)
                        .str.replace('.', '', regex=False)   # separador de miles con punto
                        .str.replace(',', '.', regex=False) # separador decimal con coma
                        .str.strip())
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

# 4. Convertir columnas num√©ricas que quedaron como texto (visitas, sesiones, unidades‚Ä¶)
def limpiar_numericos(df):
    for col in df.columns:
        # omite la columna de fecha y cualquier columna ya num√©rica
        if col != 'fecha' and df[col].dtype == object:
            # si la columna contiene letras, saltamos la conversi√≥n (ej. ASIN, t√≠tulos)
            if df[col].str.contains(r'[A-Za-z]', na=False).any():
                continue
            # si contiene d√≠gitos, eliminamos separadores de miles y convertimos
            if df[col].str.contains(r'\d', na=False).any():
                df[col] = (df[col]
                           .str.replace('.', '', regex=False)   # separador de miles con punto
                           .str.replace(',', '', regex=False)   # separador de miles con coma
                           .str.strip())
                df[col] = df[col].replace('', np.nan)
                df[col] = pd.to_numeric(df[col], errors='coerce')
    return df


# 5. Funci√≥n general que aplica todas las limpiezas y convierte la fecha
def preprocess_df(df, date_col=None):
    df = limpiar_columnas(df.copy())
    df = limpiar_porcentajes(df)
    df = limpiar_monetarios(df)
    df = limpiar_numericos(df)
    if date_col and date_col in df.columns:
        df[date_col] = pd.to_datetime(df[date_col], dayfirst=True, errors='coerce')
    return df



```

```python
# Limpiar la columna 'breakdown_incentive_fee_amount' para extraer el valor de FBA_NEW_SELECTION
def extraer_incentive_amount(texto):
    """
    Extrae valores num√©ricos de diferentes patrones en breakdown_incentive_fee_amount
    """
    if pd.isna(texto) or texto == '':
        return 0.0
    
    texto_str = str(texto)
    
    # Buscar patr√≥n FBA_NEW_SELECTION:XX.XX
    match_fba = re.search(r'FBA_NEW_SELECTION[:\s]*(\d+[.,]\d+)', texto_str)
    if match_fba:
        valor = match_fba.group(1).replace(',', '.')
        return float(valor)
    
    # Buscar cualquier n√∫mero decimal en el texto
    match_num = re.search(r'(\d+[.,]\d+)', texto_str)
    if match_num:
        valor = match_num.group(1).replace(',', '.')
        return float(valor)
    
    # Si no encuentra nada, devolver 0
    return 0.0

# Aplicar la funci√≥n mejorada
print("üîç VALORES √öNICOS ANTES DE LIMPIAR:")
print(df_tarifas_almacenamiento['breakdown_incentive_fee_amount'].unique())

df_tarifas_almacenamiento['breakdown_incentive_fee_amount_clean'] = (
    df_tarifas_almacenamiento['breakdown_incentive_fee_amount']
    .apply(extraer_incentive_amount)
)

# Verificar el resultado
print("\nüîç COMPARACI√ìN ANTES Y DESPU√âS:")
comparison = df_tarifas_almacenamiento[['breakdown_incentive_fee_amount', 'breakdown_incentive_fee_amount_clean']].head(10)
print(comparison)

print(f"\nüìä ESTAD√çSTICAS DE breakdown_incentive_fee_amount_clean:")
print(df_tarifas_almacenamiento['breakdown_incentive_fee_amount_clean'].describe())

print(f"\nüìä VALORES √öNICOS EN COLUMNA LIMPIA:")
print(df_tarifas_almacenamiento['breakdown_incentive_fee_amount_clean'].unique())
```

```python
import pandas as pd
import unicodedata
import re

def limpiar_df_envios(df, columnas_fecha=None):
    """
    Funci√≥n completa para limpiar df_envios:
    - Normaliza nombres de columnas
    - Elimina columnas completamente vac√≠as
    - Elimina filas duplicadas
    - Procesa fechas si se especifican
    """
    # Hacer una copia para no modificar el original
    df = df.copy()
    
    # Normalizar nombres de columnas
    def limpiar_columna(col):
        col = unicodedata.normalize('NFKD', col).encode('ascii', errors='ignore').decode('utf-8')
        col = re.sub(r"[\s\-:()\t]+", "_", col)
        return col.strip("_").lower()

    # Aplicar limpieza de columnas
    df.columns = [limpiar_columna(col) for col in df.columns]
    
    # üóëÔ∏è MOSTRAR COLUMNAS QUE SE VAN A ELIMINAR
    print("üîç DETECTANDO COLUMNAS COMPLETAMENTE VAC√çAS:")
    columnas_antes = df.shape[1]
    
    # Identificar columnas completamente nulas
    columnas_vacias = df.columns[df.isnull().all()].tolist()
    if columnas_vacias:
        print(f"   üìã Columnas completamente vac√≠as ({len(columnas_vacias)}):")
        for col in columnas_vacias:
            print(f"      - {col}")
    else:
        print("   ‚úÖ No hay columnas completamente vac√≠as")
    
    # Eliminar columnas con todos nulos
    df = df.dropna(axis=1, how='all')
    
    # üóëÔ∏è MOSTRAR FILAS DUPLICADAS QUE SE VAN A ELIMINAR
    print(f"\nüîç DETECTANDO FILAS DUPLICADAS:")
    filas_antes = df.shape[0]
    duplicadas = df.duplicated().sum()
    
    if duplicadas > 0:
        print(f"   üìã Filas duplicadas encontradas: {duplicadas}")
    else:
        print("   ‚úÖ No hay filas duplicadas")
    
    # Eliminar filas duplicadas
    df = df.drop_duplicates()
    
    # üìä RESUMEN DE LIMPIEZA
    columnas_despues = df.shape[1]
    filas_despues = df.shape[0]
    
    print(f"\nüìä RESUMEN DE LIMPIEZA:")
    print(f"   üìã Columnas: {columnas_antes} ‚Üí {columnas_despues} (eliminadas: {columnas_antes - columnas_despues})")
    print(f"   üìã Filas: {filas_antes} ‚Üí {filas_despues} (eliminadas: {filas_antes - filas_despues})")
    
    # Procesar fechas si se especifican
    if columnas_fecha:
        for col_fecha in columnas_fecha:
            if col_fecha in df.columns:
                print(f"\nüìÖ Procesando fecha: {col_fecha}")
                df[col_fecha] = pd.to_datetime(df[col_fecha], errors='coerce')
                print(f"   ‚úÖ Convertida a datetime")
            else:
                print(f"   ‚ö†Ô∏è Columna '{col_fecha}' no encontrada")
    
    return df

# üîß APLICAR LA FUNCI√ìN CORREGIDA A df_envios
print("üîß APLICANDO LIMPIEZA COMPLETA A df_envios")
print("=" * 60)

if not df_envios.empty:
    # Mostrar estado inicial
    print(f"üìä ESTADO INICIAL:")
    print(f"   Shape: {df_envios.shape}")
    print(f"   Columnas con todos nulos: {df_envios.isnull().all().sum()}")
    print(f"   Filas duplicadas: {df_envios.duplicated().sum()}")
    
    # Aplicar limpieza
    df_envios_limpio = limpiar_df_envios(df_envios)
    
    # Actualizar el DataFrame original
    df_envios = df_envios_limpio
    
    print(f"\n‚úÖ LIMPIEZA COMPLETADA")
    print(f"üìä ESTADO FINAL:")
    df_envios.info()
    
    print(f"\nüìã PRIMERAS FILAS DEL DATAFRAME LIMPIO:")
    display(df_envios.head(3))
    
else:
    print("‚ö†Ô∏è df_envios est√° vac√≠o - no se puede limpiar")
    
```

```python
import re
import unicodedata
from dateutil import parser
import pandas as pd

# --- Limpieza de nombres de columnas ---
def limpiar_columnas(df):
    nuevas_columnas = []
    for col in df.columns:
        col_normalizada = unicodedata.normalize('NFKD', str(col)).encode('ascii', errors='ignore').decode('utf-8')
        col_limpia = re.sub(r'[-:()\t]', ' ', col_normalizada)
        col_limpia = re.sub(r'\s+', ' ', col_limpia).strip()
        col_snake = col_limpia.replace(' ', '_').lower()
        nuevas_columnas.append(col_snake)
    df.columns = nuevas_columnas
    return df

# --- Conversi√≥n de columnas de fecha ---
def convertir_fechas(df):
    columnas_fecha = [col for col in df.columns if 'date' in col or 'date_and_time' in col]
    for col in columnas_fecha:
        try:
            df[col] = df[col].apply(lambda x: parser.parse(x) if isinstance(x, str) and pd.notnull(x) else x)
            df[col] = pd.to_datetime(df[col], utc=True, errors='coerce')
            print(f"‚úÖ '{col}' convertida a datetime")
        except Exception as e:
            print(f"‚ö†Ô∏è Error en '{col}': {e}")
    return df

# --- Aplicar al DataFrame ---
df_libro_mayor = limpiar_columnas(df_libro_mayor)
df_libro_mayor = convertir_fechas(df_libro_mayor)
```

```python
def interpretar_codigos_reason(df, columna='reason'):
    # Diccionario de traducci√≥n
    codigos = {
        'M': 'P√©rdida (ubicaci√≥n desconocida)',
        'N': 'Inventario encontrado',
        'F': 'Reconciliaci√≥n positiva',
        'Q': 'Cambio de disposici√≥n',
        'P': 'Cambio de disposici√≥n (complementario)',
        '3': 'Redefinici√≥n de SKU',
        '4': 'Redefinici√≥n de SKU inversa',
        'D': 'Inventario eliminado',
        None: 'Sin motivo',
        float('0'): 'Sin motivo',
        '0': 'Otro',  # A√±adido para cubrir el caso de '0' como string
        0: 'Otro',    # A√±adido para cubrir el caso de 0 como int
    }

    # Categor√≠as agrupadas
    categorias = {
        'P√©rdida': ['M'],
        'Reconciliaci√≥n': ['N', 'F'],
        'Cambio de disposici√≥n': ['Q', 'P'],
        'Redefinici√≥n de producto': ['3', '4'],
        'Eliminaci√≥n': ['D'],
        'Sin informaci√≥n': [None, float('0')],
    }

    # Mapear a texto
    df['reason_descripcion'] = df[columna].map(codigos)

    # Si reason es '0' y reason_descripcion es NaN, poner 'Otro'
    df.loc[(df[columna] == '0') & (df['reason_descripcion'].isnull()), 'reason_descripcion'] = 'Otro'
    df.loc[(df[columna] == 0) & (df['reason_descripcion'].isnull()), 'reason_descripcion'] = 'Otro'

    # Asignar grupo
    def categorizar(codigo):
        for categoria, codigos_lista in categorias.items():
            if codigo in codigos_lista:
                return categoria
        return 'Otro'

    df['reason_categoria'] = df[columna].apply(categorizar)

    return df
```
